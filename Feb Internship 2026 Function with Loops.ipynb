Problem 1: Mobile Recharge Validation System

```python
# Problem 1: Mobile Recharge Validation System

# Valid recharge plans
valid_plans = [199, 299, 399, 599]

def validate_recharge(amount):
    # Check if amount meets minimum requirement
    if amount < 50:
        return "Invalid: Recharge amount must be at least ₹50"
    # Check if amount is a valid plan
    elif amount not in valid_plans:
        return f"Invalid: ₹{amount} is not a valid plan. Choose from {valid_plans}"
    else:
        return f"Success: Recharge of ₹{amount} is valid!"

# While loop to allow retry
while True:
    amount = int(input("Enter recharge amount: "))
    result = validate_recharge(amount)
    print(result)

    # Stop loop if valid
    if "Success" in result:
        break
    else:
        retry = input("Try again? (yes/no): ")
        if retry.lower() != "yes":
            print("Exiting recharge system.")
            break
```

Problem 2: Inventory Reorder Alert System

```python
# Problem 2: Inventory Reorder Alert System

inventory = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 8,
    "Salt": 50,
    "Wheat": 12
}

def check_inventory(inventory):
    # Loop through each product and check stock
    for product, stock in inventory.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Stock: {stock})")

# Call the function
check_inventory(inventory)
```

Problem 3: Student Result Processing System

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    # Calculate total using loop
    total = 0
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)
    print(f"Average Marks: {average:.2f}")

    # Return Pass or Fail
    if average >= 50:
        return "Result: Pass"
    else:
        return "Result: Fail"

# Test the function
marks = [45, 67, 55, 72, 38]
result = process_result(marks)
print(result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, is_peak_hour):
    # Base fare + per km charge
    base_fare = 50
    per_km = 12
    fare = base_fare + (distance * per_km)

    # Apply 25% extra for peak hours
    if is_peak_hour:
        fare = fare * 1.25

    return fare

# While loop for retry
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    is_peak = True if peak == "yes" else False

    fare = calculate_fare(distance, is_peak)
    print(f"Estimated Cab Fare: ₹{fare:.2f}")

    retry = input("Calculate again? (yes/no): ")
    if retry.lower() != "yes":
        print("Thank you for using Cab Fare Estimator!")
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance):
    # Count present days using loop
    present = 0
    for day in attendance:
        if day == "P":
            present += 1

    total_days = len(attendance)

    # Check if total days is zero to avoid ZeroDivisionError
    if total_days == 0:
        return "Error: No attendance records found."

    # Calculate attendance percentage
    percentage = (present / total_days) * 100
    print(f"Attendance: {present}/{total_days} = {percentage:.2f}%")

    # Return eligibility
    if percentage >= 75:
        return "Status: Eligible"
    else:
        return "Status: Not Eligible"

# Take input from user
name = input("Enter employee name: ")
days = int(input("Enter total number of days: "))

# Validate days input
if days <= 0:
    print("Error: Number of days must be greater than 0.")
else:
    attendance = []

    # Take P or A input for each day
    for i in range(1, days + 1):
        status = input(f"Day {i} - Enter P (Present) or A (Absent): ").upper()
        attendance.append(status)

    print(f"\nEmployee: {name}")
    print(f"Attendance Record: {attendance}")

    result = check_attendance(attendance)
    print(result)
```

Problem 6: Password Strength Checker

```python
# Problem 6: Password Strength Checker

def check_password_strength(password):
    special_characters = "@#$"

    # Check minimum length
    if len(password) < 8:
        return "Weak Password: Minimum 8 characters required"

    # Check for at least one digit using loop
    has_digit = False
    for char in password:
        if char.isdigit():
            has_digit = True
            break

    if not has_digit:
        return "Weak Password: Must contain at least one digit"

    # Check for at least one special character using loop
    has_special = False
    for char in password:
        if char in special_characters:
            has_special = True
            break

    if not has_special:
        return "Weak Password: Must contain at least one special character (@#$)"

    return "Strong Password: All conditions met!"

# Test the function
passwords = ["hello", "hello123", "Hello@123", "abc#1"]

for pwd in passwords:
    result = check_password_strength(pwd)
    print(f"'{pwd}' → {result}")
```
